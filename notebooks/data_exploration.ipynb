```python
df.drop(columns=["loan_id"], inplace=True)
```

```python
df.nunique()
```

```python
# numeric sanity
df = df[df["income_annum"] > 0]
df = df[df["loan_amount"] > 0]
df = df[df["loan_term"] > 0]
df = df[(df["cibil_score"] >= 300) & (df["cibil_score"] <= 900)]

# assets
asset_cols = [
    "residential_assets_value",
    "commercial_assets_value",
    "luxury_assets_value",
    "bank_asset_value"
]
df = df[df[asset_cols].ge(0).all(axis=1)]
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
cat_cols = ["education","self_employed"]
```

```python
for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df
```

```python
df = df[df["income_annum"] > 0]
df = df[df["loan_amount"] > 0]
df = df[df["loan_term"] > 0]
```

```python
df = df[(df["cibil_score"] >= 300) & (df["cibil_score"] <= 900)]
```

```python
asset_cols = [
    "residential_assets_value",
    "commercial_assets_value",
    "luxury_assets_value",
    "bank_asset_value"
]

df = df[df[asset_cols].ge(0).all(axis=1)]
```

```python
df["no_of_dependents"] = df["no_of_dependents"].astype(int)
df["loan_term"] = df["loan_term"].astype(int)
```
